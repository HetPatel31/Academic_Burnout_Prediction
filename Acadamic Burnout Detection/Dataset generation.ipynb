```python
import pandas as pd
import numpy as np

# For reproducibility
np.random.seed(42)

data = []

# Locked parameters
NUM_STUDENTS = 400
NUM_WEEKS = 12

for student_id in range(1, NUM_STUDENTS + 1):

    # 0 = Normal, 1 = At-Risk
    burnout_risk = np.random.choice([0, 1], p=[0.7, 0.3])

    # Base academic level for each student
    base_quiz = np.random.normal(75, 5)
    base_attendance = np.random.normal(90, 4)

    for week in range(1, NUM_WEEKS + 1):

        if burnout_risk == 0:
            # Normal student behavior
            avg_delay = np.random.normal(0.5, 0.3)
            quiz_score = np.random.normal(base_quiz, 4)
            attendance = np.random.normal(base_attendance, 3)
            study_sessions = np.random.randint(4, 7)

        else:
            # At-risk student behavior (degrades over time)
            avg_delay = np.random.normal(1.5 + 0.15 * week, 0.4)
            quiz_score = np.random.normal(base_quiz - 0.8 * week, 5)
            attendance = np.random.normal(base_attendance - 0.6 * week, 5)
            study_sessions = np.random.randint(1, 4)

        data.append([
            student_id,
            week,
            round(max(0, avg_delay), 2),
            round(min(max(0, quiz_score), 100), 2),
            round(min(max(0, attendance), 100), 2),
            study_sessions,
            np.random.randint(1, 4),   # deadline load
            burnout_risk
        ])

# Create DataFrame
df = pd.DataFrame(data, columns=[
    "student_id",
    "week",
    "avg_submission_delay",
    "quiz_score",
    "attendance_pct",
    "study_sessions",
    "deadline_load",
    "burnout_risk"
])

# Save dataset
df.to_csv("synthetic_student_burnout_data.csv", index=False)

print("Dataset created successfully!")
print("Shape:", df.shape)

df.head()
```
